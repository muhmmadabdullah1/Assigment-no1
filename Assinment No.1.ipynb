### 39. Bill Calculator: Agar shopping 5000 se upar hai toh 20% discount, 2000-5000 par 10% discount, warna No discount.

```python
amount = float(input("Shopping amount likho: "))

if amount > 5000:
    discount = amount * 20 / 100
    final = amount - discount
    print("20% discount mila")
    print(final)
elif amount >= 2000:
    discount = amount * 10 / 100
    final = amount - discount
    print("10% discount mila")
    print(final)
else:
    print("No discount")
    print(amount)
```

### 41. Check karein ke ek saal (year) Leap Year hai ya nahi.

```python
year = int(input("Saal likho: "))

if year % 4 == 0:
    if year % 100 == 0:
        if year % 400 == 0:
            print("Leap Year")
        else:
            print("Not Leap Year")
    else:
        print("Leap Year")
else:
    print("Not Leap Year")
```

### 43. Ek simple calculator banayein (+, -, *, /) symbols ke basis par operation perform karein.

```python
num1 = float(input("Pehla number: "))
num2 = float(input("Doosra number: "))
op = input("Operation likho (+, -, *, /): ")

if op == "+":
    print(num1 + num2)
elif op == "-":
    print(num1 - num2)
elif op == "*":
    print(num1 * num2)
elif op == "/":
    if num2 == 0:
        print("Zero se divide nahi kar sakte")
    else:
        print(num1 / num2)
else:
    print("Galat operation")
```

### 44. User ka weight aur height le kar BMI category batayein (Underweight, Normal, Overweight).

```python
weight = float(input("Weight in kg likho: "))
height = float(input("Height in meters likho: "))

bmi = weight / (height * height)

if bmi < 18.5:
    print("Underweight")
elif bmi >= 18.5 and bmi <= 24.9:
    print("Normal")
else:
    print("Overweight")
```

### 45. ATM Machine Logic: Pehle PIN check karein, agar sahi ho toh Amount poochiye, agar account balance se kam amount ho toh 'Withdrawal Success'.
